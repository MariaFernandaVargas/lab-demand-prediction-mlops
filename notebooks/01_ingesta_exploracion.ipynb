```python
# 01_ingesta_exploracion.ipynb

# üì¶ Librer√≠as
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# üì• Cargar los datos
df = pd.read_csv('../data/ventas_simuladas.csv', parse_dates=['fecha'])

# üëÄ Vista r√°pida
print("Dimensiones del dataset:", df.shape)
print("\nPrimeras filas:")
display(df.head())

# ‚ÑπÔ∏è Informaci√≥n general
print("\nTipos de datos:")
print(df.dtypes)

# üßπ Valores nulos
print("\nValores nulos por columna:")
print(df.isnull().sum())

# üìä Estad√≠sticas generales
print("\nEstad√≠sticas descriptivas:")
display(df.describe())

# üìÜ Agregaci√≥n por fecha
demanda_por_fecha = df.groupby('fecha')['demanda'].sum().reset_index()

# üìà Visualizaci√≥n de la demanda total en el tiempo
plt.figure(figsize=(14, 5))
sns.lineplot(data=demanda_por_fecha, x='fecha', y='demanda')
plt.title("Demanda total diaria")
plt.xlabel("Fecha")
plt.ylabel("Demanda")
plt.grid(True)
plt.show()

# üìä Distribuci√≥n de demanda por producto
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='producto', y='demanda')
plt.title("Distribuci√≥n de demanda por producto")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
